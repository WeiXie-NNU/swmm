```python
from pyswmm import Simulation, Nodes, Links, Subcatchments

# 加载并运行模型
with Simulation('example.inp') as sim:
    nodes = Nodes(sim)
    links = Links(sim)
    subs = Subcatchments(sim)
    
    # 示例：访问一个节点属性
    j = nodes["J4"]
    print("Invert elevation:", j.invert_elevation)
    
    # 示例：访问一条管道的属性
    c = links["C3"]
    print("Outlet node:", c.outlet_node)
    
    # 时间步迭代输出
    for step in sim:
        print(sim.current_time, round(sim.percent_complete * 100, 1), "%", j.depth, c.flow)
```
